```python
# Arithmatic Operators
a= 10
b=3
print("Arithmatic Operators:")
print("Addition:",a + b)
print("Subtraction:",a - b)
print("Multiplication:",a * b)
print("Division:",a / b)
print("Floor Division:",a // b)
print("Modulo:",a % b)
print("Exponentiation:",a ** b)
print("Floor Division:", a // b)
print()
# Assignment Operators
print("Assignment Operators:")
x = 5
x += 2
print("x += 2:", x)
x -= 1
print("x -= 1:", x)
x *= 3
print("x *= 3:", x)
x /= 2
print("x /= 2:", x)
x %= 4
print("x %= 2:, x")
x **= 2
print("x **= 2:", x)
x //= 3
print("x //= 3:", x)
print()
# Comparison Operators
print("Comparison Operators:")
print("Equal to:", a == b)
print("Not equal to:", a != b)
print("Greater than:", a > b)
print("Less than:", a < b)
print("Greater than or equal to:", a >= b)
print("Less than or equal to:", a <= b)
print()
# Logical Operators
print("Logical Operators:")
print("Logical AND:", a > b and a < 15)
print("Logical OR:", a > b or a < 5)
print("Logical NOT:", not (a > b))
# Identity Operators
print("Identity Operators:")
print("is:", a is b)
print("is not:", a is not b)
print()
# Membership Operators
print("Membership Operators:")
my_list = [1, 2, 3, 4, 5]
print("in:", 3 in my_list)
print("not in:", 6 not in my_list)
tuple1 = (1, 2, 3, 4, 5)
print("in:", 3 in tuple1)
print("not in:", 6 not in tuple1)
set1 = {1, 2, 3, 4, 5}
print("in:", 3 in set1)
print("not in:", 6 not in set1)
dict1 = {'a': 1, 'b': 2, 'c': 3}
print("in:", 'a' in dict1)
print("not in:", 'd' not in dict1)
print()
# Bitwise Operators
print("Bitwise Operators:")
print("Bitwise AND:", a & b)
print("Bitwise OR:", a | b)
print("Bitwise XOR:", a ^ b)
print("Bitwise NOT:", ~a)
print("Left Shift:", a << 2)
print("Right Shift:", a >> 2)

```
